```python
#creating subset of df with only first million rows
df = ords_prods_merge[:1000000]
```

```python
#creating function to define price range of products
def price_label(row):

#if the value within the prices column within the given row is less than or equal to 5, then return the string 'low-range product'   
  if row['prices'] <= 5:
    return 'Low-range product'
#or else, if the value is greater than 5 and less than or equal to 15, then return the string 'mid-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
#or else, if the value is greater than 15, then return the string 'high-range product'
  elif row['prices'] > 15:
    return 'High range'
#or else, return the string 'not enough data'
  else: return 'Not enough data'
```

```python
#creating a column on the df called price_range as the location for the label using the new function
df['price_range']=df.apply(price_label,axis=1)
#axis = 1 tells Python to apply the function to all rows within the df
```

```python
#2.create a new column with the info on the busiest days
ords_prods_merge['busiest_days']=result_2
```

```python
#3.Check values
ords_prods_merge['busiest_days'].value_counts(dropna=False)
```

#3.Observations: This classification returns a more balanced frequency between the different options. This can have pros and cons - we might see less of a distinction between days considered busier and the ones considered slower. On the other hand it allows us for a fairer comparison between the labels, as looking at 2 vs 2 vs 3 days (in the second version of the column) provides a better comparison than looking at two individual days vs 5 days labelled as "regularly busy".

```python
#4.checking what the busiest hours are
ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

```python
#4.Creating a conditional loop for the periods of time
#empty list/shell
result_3=[]

#splitting the hours into 3 equal-sized blocks of 8 hours
for value in ords_prods_merge["order_hour_of_day"]:
  if value >=9 and value <=16:
    result_3.append("Most orders")
#most orders are placed in the period between 9 in the morning and 16 (4 in the afternoon)
  elif value <= 6 or value == 23:
    result_3.append("Fewest orders")
#the least amount of orders are placed between 23 (11 in the evening) and 6 in the morning
  else:
    result_3.append("Average orders")
#the remaining 8 hours have an average amount of orders
```

```python
#4.create a new column with the info on the busiest days
ords_prods_merge['busiest_period_of_day']=result_3
```

```python
#5.Print frequency for the new column
ords_prods_merge['busiest_period_of_day'].value_counts(dropna=False)
```
